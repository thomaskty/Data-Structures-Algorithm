```python
import os

def find_ipynb_files(root_dir, exclude_files):
    ipynb_files = []
    # Traverse through all directories and files in root_dir
    for subdir, dirs, files in os.walk(root_dir):
        for file in files:
            # Check if the file extension is .ipynb
            if file.endswith('.ipynb'):
                full_path = os.path.join(subdir, file)
                # Exclude files in the exclude list
                if file not in exclude_files:
                    ipynb_files.append(full_path)
    
    # Sort files based on the numeric prefix in the filename
    ipynb_files.sort(key=lambda x: int(x.split(os.sep)[-1].split('-')[0]))
    return ipynb_files

# Set the root directory to search for ipynb files
root_directory = './'
# Define the list of files to exclude
exclude_list = ['audit.ipynb']

sorted_ipynb_files = find_ipynb_files(root_directory, exclude_list)

# Print sorted list of ipynb files
for file in sorted_ipynb_files:
    print(file)
```
