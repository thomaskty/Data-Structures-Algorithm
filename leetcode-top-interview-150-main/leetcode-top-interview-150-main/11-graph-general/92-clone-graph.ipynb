# **Question 90 - 133. Clone Graph**

**Difficulty:** Medium

**Topics:** Depth-First Search, Breadth-First Search, Graph, Hash Table

**Description**

The challenge is to create a deep copy of a connected undirected graph. You're provided with a reference to a node within the graph. Each node has the following properties:

- **val:** An integer value representing the node.
- **neighbors:** A list of neighboring nodes (List[Node]).

**Input Format**

The input is an adjacency list (a list of lists) representing the graph. To maintain simplicity, node values correspond to their index in the adjacency list (1-indexed).

**Output**

Return a reference to the cloned graph's root node.

**Examples**

**Example 1:**

```
Input: adjList = [[2,4],[1,3],[2,4],[1,3]]
Output: [[2,4],[1,3],[2,4],[1,3]]
Explanation: (See problem description for detailed explanation)
```

**Example 2:**

```
Input: adjList = [[]]
Output: [[]]
Explanation:  A single node with no neighbors.
```

**Example 3:**

```
Input: adjList = []
Output: []
Explanation: An empty graph.
```

**Constraints**

- 0 <= Number of nodes <= 100
- 1 <= Node.val <= 100
- Node.val is unique.
- The graph is connected (all nodes reachable from the given node).
- No self-loops or repeated edges exist.

````python
# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors: list["Node"] | None = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
````

````python
from typing import Optional


class Solution:
    def cloneGraph(self, node: Optional[Node]) -> Optional[Node]:
        # hash map to store the original node and the cloned node
        visited: dict[Node, Node] = {}

        def dfs(node: Optional[Node]) -> Optional[Node]:
            # Base case: handle null input
            if not node:
                return None

            # Return existing copy if node already cloned
            if node in visited:
                return visited[node]

            # create a new node
            clone_node = Node(node.val)

            # Mark the copy as visited
            visited[node] = clone_node

            # Recursively clone neighbors and link them
            for neighbor in node.neighbors:
                clone_node.neighbors.append(dfs(neighbor))

            return clone_node

        # Start cloning from the input node
        return dfs(node)
````
