# Question 21 - 151. Reverse Words in a String

_Difficulty_: Medium

### Topics

- String Manipulation
- Data Structures

### Problem Statement

Given a string `s`, the task is to reverse the order of the words.

**Definition**: A word is a sequence of non-space characters. Words in `s` are separated by at least one space.

**Goal**: Return a string with the words in reverse order, concatenated by a single space.

**Note**: The input string `s` may have leading or trailing spaces or multiple spaces between two words. The output should have only a single space separating the words and no extra spaces.

### Examples

1. **Input**: `s = "the sky is blue"`  
   **Output**: `"blue is sky the"`

2. **Input**: `s = "  hello world  "`  
   **Output**: `"world hello"`  
   **Explanation**: The reversed string should not contain leading or trailing spaces.

3. **Input**: `s = "a good   example"`  
   **Output**: `"example good a"`  
   **Explanation**: Multiple spaces between two words should be reduced to a single space in the reversed string.

### Constraints

- `1 <= s.length <= 104`
- `s` contains English letters (upper-case and lower-case), digits, and spaces `' '`.
- There is at least one word in `s`.

### Follow-up

If the string data type is mutable in your language, can you solve it in-place with O(1) extra space?


```python
class Solution:
    def reverseWords(self, s: str) -> str:
        # convert the words into a list of characters
        chars = list(s)

        # reverse the characters in the entire list
        self.reverseCharacters(chars, 0, len(chars) - 1)

        # now we'll make the words forward again by reversing each word's characters

        # we hold the index of the *start* of the current word as we look for the end of the current word
        current_word_start_index = 0

        for i in range(len(chars) + 1):
            # found the end of the current word!
            if (i == len(chars)) or (chars[i] == " "):
                # if we haven't exhausted the string our next word's start is one character ahead
                self.reverseCharacters(chars, current_word_start_index, i - 1)
                current_word_start_index = i + 1

        # remove leading, trailing and multiple spaces
        return self.trimSpaces(chars)

    def reverseCharacters(
        self, chars: list[str], left_index: int, right_index: int
    ) -> list[str]:
        # walk towards the middle, from both sides
        while left_index < right_index:
            # swap the left and right characters
            chars[left_index], chars[right_index] = (
                chars[right_index],
                chars[left_index],
            )
            # move towards the middle
            left_index += 1
            right_index -= 1

        return chars

    def trimSpaces(self, chars: list[str]) -> str:
        # remove leading, trailing and duplicate spaces
        new_chars = []
        in_word = False

        for char in chars:
            if char != " ":
                new_chars.append(char)
                in_word = True
            elif in_word:
                # add a space between words
                new_chars.append(char)
                in_word = False

        # remove trailing spaces, if any
        if new_chars and new_chars[-1] == " ":
            new_chars.pop()

        return "".join(new_chars)

    def reverse_words(self, s: str) -> str:
        return " ".join(reversed(s.split()))
        # return ' '.join(s.split()[::-1])
```
