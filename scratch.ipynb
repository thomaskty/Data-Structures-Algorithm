```python
# first and last occurence of a sorted array 
def find_first_occurence(ar,value,first =True):
    start,end = 0,len(ar)-1
    result = -1 
    while start<=end:
        mid = int((start+end)/2)
        if value == ar[mid]:
            if first == True:
                end = mid -1
            else:
                start = mid+1 
            result = mid 
        elif value <ar[mid]:
            end = mid -1 
        elif value >ar[mid]:
            start = mid+1 
    return result
```

## Rotating and reversing arrays

```python
def reverse(ar):
    l,r = 0,len(ar)-1
    while l<r:
        ar[r],ar[l] = ar[l],ar[r]
        l+=1
        r-=1
    return ar
    
def rotate_array(ar,k):
    si = len(ar)-k 
    right,left = ar[:si],ar[si:]
    return left+right
```

```python
a = [12,39,13,8,1,0,2,1,0,0]
reverse(a)
```

```python
rotate_array(a,k = 2)
```

```python
def rotate_array_part2(ar,k):
    # reversing the array first 
    ar = reverse(ar)
    left,right = reverse(ar[:k]),reverse(ar[k:])
    return left+right 
    
```

```python
ar = [0, 0, 1, 2, 0, 1, 8, 13, 39, 12]
rotate_array_part2(ar,k=2)
```

```python
def reverse_inplace(ar,l,r):
    while l<r:
        ar[r],ar[l] = ar[l],ar[r]
        l+=1
        r-=1
```

```python
 def rotate_array_inplace(ar,k):
    reverse_inplace(ar,l=0,r = len(ar)-1)
    reverse_inplace(ar,l=0,r = k-1)
    reverse_inplace(ar,l=k,r = len(ar)-1)
     
```

```python
array = [4,5,2,6,7]
reverse_inplace(array,0,len(array)-1)
array
```

## Minimum and Maximum 

```python
def minimum(ar):
    min_value = ar[0]
    max_value = ar[0]
    for i in range(0,len(ar)):
        if min_value>=ar[i]:
            min_value = ar[i]
        if max_value<=ar[i]:
            max_value = ar[i]
    return {'minimum':min_value,'maximum':max_value}
```

```python
minimum([913,482,478,27,81,39])
```

## Counter

```python
def counter_list(ar):
    output_dict = {}
    for i in ar:
        if i not in output_dict.keys():
            output_dict[i]= 0
        output_dict[i]+=1
    return output_dict
```

```python
counter_list([1,3,3,2,8,8,8,8,2,1,0])
```

## Factorial 

```python
def factorial(n):
    if n<=1:
        return 1
    else:
        return factorial(n-1)*n
```

```python
for i in range(10):
    print(factorial(i),end='  ')
```

## Fibanocci recursion and while

```python
# recursive implementation of fibanocci series 
def fibanocci_recursion(n):
    if n==0:
        return 0 
    elif n==1:
        return 1 
    else:
        return fibanocci_recursion(n-1)+fibanocci_recursion(n-2)
```

```python
# testing the series 
for j in range(10):
    print(fibanocci_recursion(j),end = '  ')
```

```python
def fibanocci_while(n):
    if n==0:
        return 0 
    elif n==1:
        return 1 
    else:
        a,b = 0,1
        i = 0
        while i< n-1:
            next_value = a+b 
            a,b = b,next_value
            i+=1 
        return b
```

```python
for j in range(10):
    print(fibanocci_while(j),end = '  ')
```

## Prime Numbers

```python
def prime(value):
    prime = True
    check_list = list(range(2,value//2+1))
    print(check_list)
    for i in check_list:
        if value%i==0:
            prime = False
            break
    return prime
```

```python
def isprime(n):
    prime = True
    if n>1:
        half_numbers = range(2,int(n/2)+1)
        for i in half_numbers:
            if n%i ==0:
                prime = False
                break
        return prime
    else:
        return False
```

```python
print('first 100 prime numbers ... ')
for i in range(1,50):
    if isprime(i):
        print(i,end=',')
print('\n') 
```

```python
def is_prime(n, i):
    if n <= 1:
        return False
    if i == 1:
        return True
    if n % i == 0:
        return False
    return is_prime(n, i - 1)
```

```python
for i in range(0,50):
    if is_prime(i,i-1):
        print(i,end=',')
```

## Reversing and rotation 

```python
# reversing an array using pointers 
def reverse(ar):
    i = 0 
    j = len(ar)-1
    while i<=len(ar)//2:
        ar[i],ar[j] = ar[j],ar[i]
        i+=1
        j+=1
    return ar 
```

```python
# reversing an array based on recursion 
def reverse_recur(ar):
    if len(ar)<=1:
        return ar 
    else:
        return reverse_recur(ar[1:])+reverse_recur(ar[:1])
```
